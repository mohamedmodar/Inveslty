```python
import pandas as pd
import glob
import os
import json

folder_path = '/content/drive/MyDrive/GRAD'

json_files = glob.glob(os.path.join(folder_path, '*.json'))

# df_list = [pd.read_json(file) for file in json_files]
# combined_df = pd.concat(df_list, ignore_index=True)

# print(combined_df.head())

print(json_files)
```

```python
df_list = []

for path in json_files:
    with open(path, 'r') as f:
        data = json.load(f)
    area_name = os.path.basename(path).replace('.json', '')
    df = pd.DataFrame([data])
    df['area'] = area_name
    df_list.append(df)
```

```python
combined_df = pd.concat(df_list, ignore_index=True)
combined_df
```

```python
import re

def clean_area_name(area):
    # Remove '-params'
    area = area.replace('-params', '')
    # Remove things like ' (1)', ' (2)', etc.
    area = re.sub(r'\s*\(\d+\)', '', area)
    # Replace hyphens with spaces
    area = area.replace('-', ' ')
    return area.strip()

combined_df['area'] = combined_df['area'].apply(clean_area_name)
```

```python
cols = ['area'] + [col for col in combined_df.columns if col != 'area']
combined_df = combined_df[cols]

cols = [col for col in combined_df.columns if col != 'R2'] + ["R2"]
combined_df = combined_df[cols]
```

```python
combined_df = combined_df[combined_df["area"] != "alexandria compounds"]
```

```python
combined_df.to_csv("Areas_Accuracy.csv")
```
